- Using mannwhitney test for B&C and A&C as one variable is normal while the other is not normally distributed.

```python
is_no_difference_A_B = ttest_ind(A,B)[1]>=0.05

# Is wrong cannot use ttest with One normal and other non-normal distribution
is_no_difference_B_C = ttest_ind(B,C)[1]>=0.05
is_no_difference_A_C = ttest_ind(A,C)[1]>=0.05

print("Not statistically significantly different : ",is_no_difference_A_B)

# Is wrong cannot use ttest with One normal and other non-normal distribution
print("Incorrect Result cannot use non-normal distribution B&C : ",is_no_difference_B_C)
print("Incorrect Result cannot use non-normal distribution A&C : ",is_no_difference_A_C)
```
