```python
import math
import numpy as np
from scipy.stats import shapiro, kstest 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a = np.array([170, 192, 184, 165, 176, 181, 163])

b = np.array([170, 170, 170, 170, 192, 192, 192, 192, 184, 184, 184, 184, 165, 165, 165, 165, 176, 176, 176, 176, 181, 181, 181, 181, 163, 163, 163, 163 ])
```

```python
a_log = np.log(a)
shapiro(a_log)
```

Since the p-Value is greater than 0.05 the data is normally distributed.



```python
plt.hist(a,edgecolor='black')
```

```python
b_log = np.log(b)
shapiro(b_log)
```
