```python
import math
import numpy as np
from scipy.stats import shapiro, kstest 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
from statistics import mean, median


data = pd.read_csv("./graph_data.txt",delim_whitespace=True)

print(data.head())
no_smoothing = data['No_Smoothing']

mean_no_smoothing = mean(no_smoothing)
print("Mean - No Smoothing : ",mean_no_smoothing)

media_no_smoothing = median(no_smoothing)
print("Median - No Smoothing : ",media_no_smoothing)
```

```python
sns.distplot(no_smoothing,hist=True, kde=True)
```

```python
no_smoothing_2000 = data.loc[data['Year'] >= 2000]['No_Smoothing']
```

```python
mean_no_smoothing_2000 = mean(no_smoothing_2000)
print("Mean - No Smoothing : ",mean_no_smoothing_2000)

media_no_smoothing_2000 = median(no_smoothing_2000)
print("Median - No Smoothing : ",media_no_smoothing_2000)

sns.distplot(no_smoothing_2000,hist=True, kde=True)
```
