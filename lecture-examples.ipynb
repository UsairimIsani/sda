```python
# Importing Pandas for Analysis
import pandas as pd
```

### Histogram

```python
# Setting up data and bins for the example 
hist_data = pd.Series((170,192,184,165,176,181,163))
hist_bins = 160,170,180,190,200
```

```python
# Plotting the actual histogram from the data and the bins provided.

pd.DataFrame(hist_data).plot(kind="hist",bins=hist_bins)
```

### Density Plot

```python
density_data = (170,192,184,165,176,181,163)
```

```python
pd.DataFrame(density_data).plot(kind="density")
```
